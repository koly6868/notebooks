```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn import linear_model
```

```python
x = np.arange(1,100)
y =np.vectorize(lambda x: 2*x + 5)(x)
print(x,y)
```

```python
e = np.random.normal(loc=1,scale=5, size= len(x))
Y = e + y
plt.plot(Y, 'ro')
```

```python
model = linear_model.Ridge(alpha=0.5)
X = np.reshape(x,(len(x),1) )
model.fit(X,Y)
```

```python
print(model.coef_)
```

```python
plt.plot(x,model.predict(X),Y)
```
